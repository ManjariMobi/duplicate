```python
# Balance the dataset using under-fitting random sampling
shuffled_df = df.sample(frac=1,random_state=25)
fraud_df = shuffled_df[shuffled_df["TARGET"]==1]
# Get the number of fraud cases
n_fraud = len(fraud_df)
# Sample the same number of non-fraud cases
nonfraud_df = shuffled_df[shuffled_df["TARGET"]==0].sample(n=n_fraud,random_state=25)
balanced_df = pd.concat([fraud_df,nonfraud_df])
```

```python
x=balanced_df.drop(["TARGET"],axis=1)
y=balanced_df[["TARGET"]]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=25)
```

```python
x_train.shape,x_test.shape
```

```python
y_train.shape,y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
# Find columns with object (string) dtype in x_train
object_cols = x_train.select_dtypes(include=['object']).columns

# Print the unique values in these columns to identify the 'Y' value
for col in object_cols:
    print(f"Unique values in column '{col}': {x_train[col].unique()}")

# Apply Label Encoding or One-Hot Encoding to these columns
# For example, using Label Encoding:
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for col in object_cols:
    x_train[col] = le.fit_transform(x_train[col])
    x_test[col] = le.transform(x_test[col]) # Use transform, not fit_transform for x_test

# Now, apply StandardScaler
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
#Architect the Deep learning Model
```

```python
import tensorflow as tf
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout
```

```python
model=Sequential()
model.add(Dense(256,activation="relu",input_shape=(x_train.shape[1],)))
model.add(Dropout(0.3))
model.add(Dense(128,activation="relu"))
model.add(Dropout(0.3))
model.add(Dense(64,activation="relu"))
model.add(Dropout(0.3))
model.add(Dense(1,activation="sigmoid"))
```

```python
model.summary()
```

```python
model.compile(optimizer="adam",loss="binary_crossentropy",metrics=["accuracy"])
```

```python
model.fit(x_train,y_train,validation_data=(x_test,y_test),epochs=50)
```

```python
pred=(model.predict(x_test)>0.5)*1.0
pred
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
print(confusion_matrix(pred,y_test))
```

```python
print(classification_report(pred,y_test))
```
